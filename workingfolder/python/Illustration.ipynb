```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parameters for the normal distributions
mean1, variance1 = 2, 4
mean2, variance2 = 5, 2
average_mean = (mean1 + mean2) / 2  # Compute the average of the two means

# Generating random samples from the distributions
samples1 = np.random.normal(mean1, np.sqrt(variance1), 100)
samples2 = np.random.normal(mean2, np.sqrt(variance2), 100)


# Plotting smooth distributions
x = np.linspace(-5, 10, 1000)
pdf1 = norm.pdf(x, mean1, np.sqrt(variance1))
pdf2 = norm.pdf(x, mean2, np.sqrt(variance2))


## shock 

shock = -1.5
```

```python
# Plotting histograms with fewer bins (e.g., 15 bins)
plt.hist(samples1, bins=10, alpha=0.3, label='Histogram 1', density=True)
#plt.hist(samples2, bins=10, alpha=0.3, label='Histogram 2', density=True)

plt.plot(x, pdf1, 'b-', label='Dist 1')
#plt.plot(x, pdf2, 'r-', label='Dist 2')

# Plotting means using vertical lines
plt.axvline(x=mean1, color='blue', linestyle='--', label=r'$y_{1,t+h}$')
#plt.axvline(x=mean2, color='red', linestyle='--', label='Exp 2')
#plt.axvline(x=average_mean, color='green', linestyle='--', label='Average Exp')  # Plotting the average mean

# Adding the legend
plt.legend(loc='center left',bbox_to_anchor=(1, 0.5))

# Find the current lowest y-value
ymin, ymax = plt.ylim()

# Create a margin below the graph to fit the brackets, labels, and arrows
margin = 0.3 * (ymax - ymin)
plt.ylim(ymin - margin, ymax)

# Position for the "Uncertainty of 1" bracket and label
bracket_y_position_uncertainty = ymin - 0.25 * margin
label_y_position_uncertainty = ymin - 0.8 * margin

## Position for the "Uncertainty of 2" bracket and label
#bracket_y_position_uncertainty2 = ymin - 0.35 * margin
#label_y_position_uncertainty2 = ymin - 0.8 * margin

# Adjusted "Uncertainty of 1" label position slightly to the left
adjusted_mean1_position = mean1 - 0.5

# Adjusted "Uncertainty of 2" label position slightly to the left
#adjusted_mean2_position = mean2 + 0.5


# Adding "Uncertainty of 1" bracket
plt.annotate(
    '', 
    xy=(min(samples1), bracket_y_position_uncertainty), 
    xycoords='data',
    xytext=(max(samples1), bracket_y_position_uncertainty), 
    textcoords='data',
    arrowprops={'arrowstyle': '<->'}
)

# Adding "Uncertainty of 1" label with arrow pointing to the bracket
plt.annotate(
    r'$Var_{1,t+h}$', 
    xy=(adjusted_mean1_position, bracket_y_position_uncertainty), 
    xycoords='data', 
    xytext=(adjusted_mean1_position, label_y_position_uncertainty),
    textcoords='data', 
    ha='center',
    va='center', 
    fontsize=12,
    arrowprops=dict(facecolor='black', arrowstyle='->')
)


# Adding "Uncertainty of 2" bracket
#plt.annotate(
#    '', 
#    xy=(min(samples2), bracket_y_position_uncertainty2), 
#    xycoords='data',
#    xytext=(max(samples2), bracket_y_position_uncertainty2), 
#    textcoords='data',
#    arrowprops={'arrowstyle': '<->'}
#)

# Adding "Uncertainty of 2" label with arrow pointing to the bracket
#plt.annotate(
#    'Var 2', 
#    xy=(adjusted_mean2_position, bracket_y_position_uncertainty2), 
#    xycoords='data', 
#    xytext=(adjusted_mean2_position, label_y_position_uncertainty2),
#    textcoords='data', 
#    ha='center',
#    va='center', 
#    fontsize=12,
#    arrowprops=dict(facecolor='black', arrowstyle='->')
#)


# Position for the "Disagreement" bracket and label, moved further down
#bracket_y_position_disagreement = ymin - 0.75 * margin
#label_y_position_disagreement = ymin - 1.1 * margin

## Adding "Disagreement" bracket
#plt.annotate(
#    '', 
#    xy=(mean1, bracket_y_position_disagreement), 
#    xycoords='data',
#    xytext=(mean2, bracket_y_position_disagreement), 
#    textcoords='data',
#    arrowprops={'arrowstyle': '<->'}
#)

# Adding "Disagreement" label with arrow pointing to the bracket
#plt.annotate(
#    'Disg', 
#    xy=((mean1 + mean2) / 2, bracket_y_position_disagreement), 
#    xycoords='data', 
#    xytext=((mean1 + mean2) / 2, label_y_position_disagreement),
#    textcoords='data', 
#    ha='center',
#    va='center', 
#    fontsize=12,
#    arrowprops=dict(facecolor='black', arrowstyle='->')
#)

# Labelling and showing the plot
#plt.title("Histogram and PDF of Two Normal Distributions with Means")

# Remove axis tick labels
plt.xticks([])  # Remove x-axis tick labels
plt.yticks([])  # Remove y-axis tick labels

# Remove the frame (spines) of the graph
ax = plt.gca()  # Get current Axes object
for spine in ax.spines.values():
    spine.set_visible(False)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()  # Ensures everything fits well
plt.savefig('../graphs/illustration/figure1.pdf')
plt.show()

```
